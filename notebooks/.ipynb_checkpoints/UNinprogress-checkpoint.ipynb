10. For this question, you're going to create some visualizations to compare GDP per capita values for the years 1990, 2000, 2010, and 2020. 
Start by subsetting your data to just these 4 years into a new DataFrame named gdp_decades. Using this, create the following 4 plots:
	* A boxplot
	* A barplot (check out the Barplot with Seaborn section: https://www.python-graph-gallery.com/barplot/#Seaborn)
	* A scatterplot
	* A scatterplot with a trend line overlaid (see this regplot example: https://www.python-graph-gallery.com/42-custom-linear-regression-fit-seaborn)  
Comment on what you observe has happened to GDP values over time and the relative strengths and weaknesses of each type of plot.



```python
gdp_decades = gdpdf[gdpdf['Year'].isin([1990, 2000, 2010, 2020])]
gdp_decades
```

```python
sns.boxplot(x=gdp_decades["Year"], y=gdp_decades["GDP per Capita"] )
```

```python
sns.barplot(
    x="Year", 
    y="GDP per Capita", 
    data=gdp_decades, 
    estimator=sum, 
    ci=None, 
    color='#69b3a2');
```

```python
sns.regplot(
    x=gdp_decades["Year"],
    y=gdp_decades["GDP per Capita"],
    fit_reg=False
)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.regplot(
    x=gdp_decades["Year"],
    y=gdp_decades["GDP per Capita"],
    line_kws={"color": "r"},
    ax=ax,
    
)
plt.show()
```

20. Merge the two DataFrames to one. Do this in a way that keeps **all rows** from each of the two DataFrames. Call the new DataFrame `gdp_and_internet_use`. Look at the first and last few rows to confirm that it merged correctly.

```python
gdpdf['Year'] = gdpdf['Year'].astype(str)
gdpdf

gdp_and_internet_use = pd.merge(gdpdf, internet_df, on=['Year','Country'])
gdp_and_internet_use
```
